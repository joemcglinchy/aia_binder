```python
scene_path = 'landsat-pds/c1/L8/033/044/LC08_L1TP_033044_20180519_20180605_01_T1'
base = os.path.basename(scene_path) # all after last slash
bands = ['B{}'.format(i) for i in range(1,9)]
bands = ['B1', 'B2', 'B3', 'B4', 'B5', 'B6', 'B7']

band_arrs = []
for band in bands:
    sample_band = scene_path + '/LC08_L1TP_033044_20180519_20180605_01_T1_{}.TIF'.format(band)
    s3_path = "s3://{}".format(sample_band)
    print(s3_path)
    with rasterio.open(s3_path, 'r') as src:
        arr = src.read(1) # to read in as shape [rows, cols], not [band, rows, cols]
        
    band_arrs.append(arr)
    
```

```python
#!aws s3 ls s3://landsat-pds/c1/L8/033/044/LC08_L1TP_033044_20180519_20180605_01_T1/LC08_L1TP_033044_20180519_20180605_01_T1_B9.TIF
```

```python
# convert to a numpy array

import numpy as np
l8_arr = np.array(band_arrs)
del band_arrs
```

```python
# plot the bands

from matplotlib import pyplot as plt
plt.figure(figsize=(20,10))

# plot the bands
for i in range(l8_arr.shape[0]):
    plt.subplot(2,4,i+1)
    plt.axis('off')
    plt.imshow(l8_arr[i], cmap="Greys_r")
    plt.title(bands[i])
plt.show()



              
```

```python
# plot a veg index
red = l8_arr[3,:,:]
nir = l8_arr[4,:,:]
ndvi = (nir - red) / (nir + red)

plt.imshow(ndvi)
plt.show()
```
